Fonction "resample"


Elle permet de regrouper les donnees selon une frequence.

```python
bitcoin.loc['2023', 'Close'].resample('M').mean().plot()
```

```python
bitcoin.loc['2023', 'Close'].resample('2W').mean().plot()
```

```python
bitcoin.loc['2023', 'Close'].resample('2W').std().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2023', 'Close'].plot()
bitcoin.loc['2023', 'Close'].resample('M').mean().plot(label='moyenne par mois', lw=3, ls=':',alpha = 0.8)
bitcoin.loc['2023', 'Close'].resample('W').mean().plot(label='moyenne par semaine', lw=3, ls='--',alpha = 0.8)
plt.legend()
```

Fonctions aggragates: nous permet de rassembler dans un seul tableau plusiers statistiques qu'on aimerait faire par dessus resample.

```python
m = bitcoin.loc['2023', 'Close'].resample('W').agg(['mean','std', 'min', 'max'])

plt.figure(figsize=(12,8))
m['mean']['2023'].plot(label='moyenne par semaine')
plt.fill_between(m.index, m['max'], m['min'], alpha=0.2, label ='min -max par semaine')
plt.legend()
```

Moving average : permet de faire la moyenne sur une fenetre des valeurs avec un decalage voulu ou souhaite.

```python
bitcoin.loc['2023', 'Close'].rolling(window=7)
```

```python
bitcoin.loc['2023', 'Close'].rolling(window=7).mean().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2023', 'Close'].plot()
bitcoin.loc['2023', 'Close'].rolling(window=7).mean().plot(label='moving average', lw=3,alpha=0.8)
plt.legend()
```

Exp weighted function = moyenne mobile exponentielle : les valuers perdent peu a peu du poids avec le temps

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2023', 'Close'].plot()
bitcoin.loc['2023', 'Close'].rolling(window=7).mean().plot(label='mon centre', lw=3,alpha=0.8)
bitcoin.loc['2023', 'Close'].rolling(window=7,center = True).mean().plot(label ='centre',lw=3, ls='--')
bitcoin.loc['2023', 'Close'].ewm(alpha=0.6).mean().plot(label ='centre',lw=3, ls=':')
plt.legend()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2023', 'Close'].plot()
for i in np.arange(0.2, 1, 0.2):
  bitcoin.loc['2023', 'Close'].ewm(alpha=i).mean().plot(label = f'emw {i}', ls='--', alpha = 0.8)
plt.legend()
```

```python
# @title year vs Close_btc

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['Close_btc']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = btc_eth.sort_values('year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('year')
_ = plt.ylabel('Close_btc')
```
